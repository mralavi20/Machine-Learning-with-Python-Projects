Importing libaries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

read files into dataframes

```python
movies = pd.read_csv ("movies.csv")
ratings = pd.read_csv ("ratings.csv")
```

```python
movies.head ()
```

```python
ratings.head ()
```

remove years from title

```python
movies["title"] = movies.title.str.replace ("(\(\d\d\d\d\))", "")
movies["title"] = movies["title"].apply (lambda x: x.strip())
movies.head ()
```

Split geners into a list

```python
movies["genres"] = movies.genres.str.split ("|")
movies.head ()
```

convert the list of genres to a vector

```python
for i, r in movies.iterrows ():
    for g in r["genres"]:
        movies.at[i, g] = 1

movies = movies.fillna (0)
movies.head ()
```

Remove timestamp from ratings dataframe

```python
ratings = ratings.drop ("timestamp", 1)
ratings.head ()
```

Create a user

```python
user = [
            {'title':'Breakfast Club, The', 'rating':4},
            {'title':'Toy Story', 'rating':4},
            {'title':'Jumanji', 'rating':3},
            {'title':"Pulp Fiction", 'rating':5},
            {'title':'Akira', 'rating':2}
         ]

input_movies = pd.DataFrame (user)
input_movies
```

Add movie ID to user

```python
input_id = movies[movies["title"].isin (input_movies["title"].tolist ())]

input_movies = pd.merge (input_id, input_movies)

input_movies
```

Get the subset of movies that user has watched

```python
user_movies = movies[movies["movieId"].isin (input_movies["movieId"].tolist ())]
user_movies
```

Reset index and drop movie id

```python
user_movies = user_movies.reset_index (drop=True)

user_genre = user_movies.drop ('movieId', 1).drop ('title', 1).drop ('genres', 1)

user_genre
```

Turn each genre into weights

```python
user_profile = user_genre.transpose ().dot (input_movies["rating"])

user_profile
```

Extract genre table from the original dataframe

```python
genre_table = movies.set_index (movies["movieId"])

genre_table = genre_table.drop ("movieId", 1).drop ("title", 1).drop ("genres", 1)

genre_table.head ()
```

Multiply the genres by the weights and take weighted average

```python
recommendation = ((genre_table * user_profile).sum (axis = 1)) / (user_profile.sum ())
recommendation.head ()
```

Sort recommendations table in descending order

```python
recommendation = recommendation.sort_values (ascending = False)

recommendation.head ()
```

Find movies name

```python
movies_name = movies.loc[movies["movieId"].isin (recommendation.head (50).keys ())]

movies_name.head (50)
```
