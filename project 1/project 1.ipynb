```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv ("data.csv")
```

```python
data1 = df[df["Area"] < 3000]
data2 = data1[pd.isna (data1["Address"]) == False]
```

```python
msk = np.random.rand (len (data2)) < 0.8
train = data2[msk]
test = data2[~msk]
```

```python
r = linear_model.LinearRegression ()
train_x = np.asanyarray (train[["Area"]])
train_y = np.asanyarray (train[["Price"]])
r.fit (train_x, train_y)
```

```python
plt.scatter (train.Area, train.Price, color = "blue")
plt.plot (train_x, r.coef_[0][0] * train_x + r.intercept_[0], "-r")
```
